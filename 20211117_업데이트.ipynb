```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd

#각각의 이름을 간편하게 as __로 바꿔서 불러오기
```

```python
# parameter setting
amp = 1          # range [0.0, 1.0]       # amplitude 진폭. wave의 높이
sr = 10000       # sampling rate, Hz      # sampling rate. digital 해상도, 초당 얼마나 잘게 쪼개는지를 Hz로 나타냄.
dur = 0.5        # in seconds             # 지속 시간
freq = 100.0     # sine frequency, Hz     # cycle이 1초에 몇 번 반복하는지.
phase = np.pi/2  # phase, radian          # 사인 곡선의 시작점. pi/2는 90도.
```

```python
# generate samples
nSamp = sr * dur                # 총 몇 개의 샘플이 필요한지
samples = np.arange(1, nSamp+1)    # 1에서 5000(nSamp값) 까지 등분한 숫자를 generate 시킴 
```

```python
# convert samples to time
t = samples/sr                  # time 샘플 만들기. 1/10000초부터 5000/10000초까지.
```

```python
# convert time to angle in radian
theta0 = t * 2*np.pi * freq       # radian으로 변환: 한 바퀴의 앵글 * 주파수 
```

```python
# shift angle by phase
theta = theta0 + phase          # 시작점이 0도가 아닐 경우 phase 더해줌
```

```python
# generate signal by cosine-phasor
s = amp * np.sin(theta)     # 사인 곡선 만들기
```

```python
# plot the entire signal
plt.plot(t, s, '.')     # t를 x축으로, s를 y축으로 plotting. '.'으로 표시(이거 안쓰면 실선으로 나옴)
plt.xlabel('time (s)')  # x축 이름 정해주기
plt.ylabel('real')      # y축 이름 정해주기
```

```python
# plot a certain of signals         # 부분을 확대해서 보는 것. 
sampleRng = np.arange(0,100)        # 0부터 100번째의 sample 볼 수 있음
plt.plot(t[sampleRng], s[sampleRng], ':')
plt.xlabel('time (s)')              # x축과 y축 정의
plt.ylabel('real')
```

```python
ipd.Audio(s, rate=10000)      # 오디오 출력. rate은 sample rate을 의미함.
```

```python
def phasor_sine(amp, sr, dur, freq, phase):       # def 이용하여 위의 단계들을 압축해서 정의해줌
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta                            # 여기에서 정의한 phasor_sine 함수에서의 출력값
```

```python
amp = 1; sr = 10000; dur = 1; freq = 440; phase =0 
s, t, theta = phasor_sine(amp, sr, dur, freq, phase)
ipd.Audio(s, rate=sr)
```
