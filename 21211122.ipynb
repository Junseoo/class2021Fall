```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
amp = 1 ; sr = 10000 ; phase = 0

D = 293.66
D# = 311.13
E = 329.63
F = 349.23
G = 392.00
A = 440.00
B = 493.88
C = 523.25
```

```python
# Jingle-Bell Rock

notes = [C, 0, C, 0, C, 0, B, 0, B, 0, B, 0, A, B, A, E, G, 
         A, B, A, F, D, E, F, G, A, G, D, E, F, G, A, G, A, G, C]

durs = [0.45, 0.05, 0.45, 0.05, 0.95, 0.05, 0.45, 0.05, 0.45, 0.05, 0.95, 0.05,
        0.5, 0.5, 1, 1, 1, 0.5, 0.5, 1, 2, 0.5, 1, 0.5, 0.5, 1, 0.5, 0.5, 0.5, 
        0.5, 2.5, 1, 0.5, 1, 1, 2.5]
```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
  s, t, th= phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all, s))
```
